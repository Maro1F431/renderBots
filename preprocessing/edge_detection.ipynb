```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('/home/marow/Work/Epita/Multiagent/final_project/renderBots/includes/data/render_test.jpg',0)
edges = cv.Canny(img,30,650)
edges = cv.bitwise_not(edges)
thresh = 128
edges = cv.threshold(edges, thresh, 255, cv.THRESH_BINARY)[1]
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
cv.imwrite("../includes/data/render_test_edges.png", edges)
```
